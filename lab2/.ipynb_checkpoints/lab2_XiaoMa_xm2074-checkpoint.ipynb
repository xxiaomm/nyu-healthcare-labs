### 1. In column “Age”, there is mix data which contains both number and characters.

```python
data[["Age", "Years"]] = data["Age"].str.split(' ', expand = True)
data.drop("Years", axis = 1, inplace = True)
data.dtypes
```

```python
data["Age"] = pd.to_numeric(data["Age"])
data.dtypes
```

```python
data
```

### 2. In column “Weight” and “Height”, there is serval empty data which is unneeded.

```python
data.dropna(axis = 0,  how='any', inplace = True)
data
```

### 3. The type of colums “Systolic_BP&Diastolic_BP” is wrong. It has two variables. It should be seperated into two colums: “Systolic_BP” and “Diastolic_BP”. 


```python
data[["Systolic_BP", "Diastolic_BP"]] = data["Systolic_BP&Diastolic_BP"].str.split("/", expand = True)
data.dtypes
```

### 4. And the two colums: “Systolic_BP” and “Diastolic_BP” should be the type of numeric.

```python
data.drop("Systolic_BP&Diastolic_BP", axis = 1, inplace = True)
data["Systolic_BP"] = pd.to_numeric(data["Systolic_BP"])
data["Diastolic_BP"] = pd.to_numeric(data["Diastolic_BP"])
data.dtypes
```

```python
# https://www.who.int/data/gho/data/indicators/indicator-details/GHO/15-19-years-old-current-drinkers-(-)
data2 = pd.read_csv("youth_current_drinkers - copy.csv")
data2
```

### 1. delete columns that all rows are empty

```python
data2.dropna(axis = 1,  how='all', inplace = True)
# data2.dropna(axis = 0,  how='how', inplace = True)
data2.info()
```

### 2. delete inneeded (duplicate information) column "Value"

```python
data2.drop("Value", axis = 1, inplace = True)
data2
```

### 3. delete column "Dim1 type" and change the column name of "Dim1" to "sex

```python
data2.drop("Dim1 type", axis = 1, inplace = True)
data2 = data2.rename(columns={"Dim1":"Sex"})
data2
```

 ### 4. use Boxplots to identify outliers of column "FactValueNumeric".

```python
plt.boxplot(data2["FactValueNumeric"])
plt.show()
```

### 5. show missing values

```python
data2.isnull().sum()
```

### 6. split date into "month", "day, "year"

```python
# data2 = pd.DataFrame({"DateModified": []})
data2[["Date", "Time"]] = data2["DateModified"].str.split(" ", expand = True)
data2.drop("DateModified", axis = 1, inplace = True)
data2.drop("Time", axis = 1, inplace = True)
data2
```

### 7.change “Date” to date format and split them into three columns "month", "day, "year"

```python
data2["Date"] = pd.to_datetime(data2["Date"])
data2["Month"] = data2["Date"].dt.month
data2["Day"] = data2["Date"].dt.day
data2["Year"] = data2["Date"].dt.year
data2
```
